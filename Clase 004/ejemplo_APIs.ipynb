API de Pokemon https://pokeapi.co/

```python

pokemon_name = "snorlax"
url = f"https://pokeapi.co/api/v2/pokemon/{pokemon_name}"

response = requests.get(url)
data = response.json()

data
```

revisar https://jsoncrack.com/

```python
# Extraer información del Pokémon
name = data["name"]
weight = data["weight"]
height = data["height"]
abilities = [ability["ability"]["name"] for ability in data["abilities"]]
stats = {stat["stat"]["name"]: stat["base_stat"] for stat in data["stats"]}


print("Nombre:", name)
print("Peso:", weight)
print("Altura:", height)
print("Habilidades:", abilities)
print("Estadísticas:", stats)
```
