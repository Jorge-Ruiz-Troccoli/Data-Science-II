Un f-string es una forma de formatear cadenas de texto en Python introducida a partir de la versión 3.6. El término "f-string" proviene del prefijo "f" que se coloca antes de la cadena entre comillas.

Los f-strings permiten la interpolación de variables y expresiones dentro de las cadenas de forma más concisa y legible. Al utilizar un f-string, puedes incluir expresiones y variables dentro de la cadena utilizando llaves {}. Estas expresiones se evalúan y se insertan en la cadena resultante.

Aquí hay un ejemplo simple que muestra cómo usar f-strings:

```python
g=9.8

print(f"el valor de aceleración de la gravedad en la tierra es {g} m/s2")
```

```python
#usando print se puede hacer lo mismo pero es menos elegante
nombre="Jorge"
ocupación="profe"

print("Hola, me llamo", nombre, "y soy", ocupación, "en CH")
```

```python
numero = 3.14159
cadena = f"El valor de PI es {numero:.1f}"
cadena
```

```python
nombres = ["Alice", "Bob", "Charlie"]
diccionario = {"nombre": "Alice", "edad": 25}

cadena1 = f"El primer nombre es {nombres[0]}"
cadena2 = f"El nombre es {diccionario['nombre']} y tiene {diccionario['edad']} años"

print(cadena1)
print(cadena2)
```

El operador %, se usaba en el formateo de cadenas en Python antes de la incorporación de los f-strings (desde la versión 3.6). Aunque aún es válido, se recomienda utilizar f-strings en su lugar, ya que proporcionan una sintaxis más legible,poderosa y "pytonesca".

```python
# uso del operador % en formateo de cadenas
nombre = "Juan"
edad = 10

cadena = "Hola, me llamo %s y tengo %d años." % (nombre, edad)
print(cadena)

# %s para una cadena
# %d para un número entero
# %.2f para un número de punto flotante.
```

# Casos duplicados:

```python
# Datos de ejemplo
data = {'Nombre': ['Juan', 'María', 'Pedro', 'Ana', 'María', 'María'],
        'Edad': [25, 30, 35, 40, 30, 15]}

edades = pd.DataFrame(data)

edades

```

```python
# Detección de duplicados
duplicados = edades.duplicated()

# Mostrar los duplicados
print("Duplicados:")
edades[duplicados]
```

```python
# Eliminación de duplicados
df_sin_duplicados = edades.drop_duplicates()

# Mostrar el DataFrame sin duplicados
print("\nDataFrame sin duplicados:")
df_sin_duplicados
```

```python
# detección basados en una columna

duplicados = edades.duplicated("Nombre")

# Mostrar los duplicados
print("Duplicados:")
edades[duplicados]
```

```python
# Eliminación de duplicados
duplicados = edades.drop_duplicates("Nombre")

# si usamos keep="last" mantenemos solamente a María de 15 años.

# Mostrar el DataFrame sin duplicados
print("\nDataFrame sin duplicados:")
duplicados
```

# Pivot table o tabla pivote

#### Indice

```python
"""
En este ejemplo se desea agrupar los datos por la columna 'sex' y utilizarla como índice de la tabla dinámica.

Por defecto, calculará la media de cada columna numérica, pero se pueden especificar otras funciones de agregación
utilizando el parámetro aggfunc
"""

pd.pivot_table(df[["tip", "total_bill", 'sex']],
               index='sex')
```

```python
#las tablas de pivote funcionan de forma similar al groupby
df[["tip", "total_bill", 'smoker']].groupby("smoker", observed=False).mean()
```

```python
df2 = pd.pivot_table(df[["tip", "total_bill", 'day', "sex"]],
               index=['sex', 'day']).round(1)

df2
```

#### Valores

```python
pd.pivot_table(df,
               index=['sex', 'smoker'],
               values=['tip'])
```

#### Agregación

```python
pd.pivot_table(df,
               index=['sex', 'smoker'],
               values=['tip'],
               aggfunc="max")
```

```python
df3 = pd.pivot_table(df,
               index=['sex', 'smoker'],
               values=['tip'],
               aggfunc=["sum", "mean", "median"])

df3 = df3.rename(columns={'sum': 'suma', 'mean': 'media'})
df3
```

```python
df3 = df.groupby(['sex', 'smoker'])['tip'].agg(['sum', 'mean'])
df3 = df3.rename(columns={'sum': 'suma', 'mean': 'media'})
df3
```

```python
df3.unstack('sex')
```

```python
df['pickup'] = pd.to_datetime(df['pickup'])
df['dropoff'] = pd.to_datetime(df['dropoff'])
```
