# SQLite

SQLite es probablemente la base de datos más sencilla para conectarse con una aplicación de Python, ya que no necesita instalar ningún módulo SQL de Python externo para hacerlo. De manera predeterminada, su instalación de Python contiene una biblioteca SQL de Python llamada sqlite3 que puede usar para interactuar con una base de datos SQLite.

Además, las bases de datos SQLite no tienen servidor y son independientes, ya que leen y escriben datos en un archivo. Esto significa que, a diferencia de MySQL y PostgreSQL, ¡ni siquiera necesita instalar y ejecutar un servidor SQLite para realizar operaciones de base de datos!

# CRUD
Es un acrónimo que se utiliza en el ámbito de las bases de datos y la programación para referirse a las operaciones básicas que se pueden realizar sobre los datos en una base de datos. CRUD significa:

Create (Crear): Permite la creación de nuevos registros o entradas en la base de datos.

Read (Leer): Permite la lectura o consulta de los registros existentes en la base de datos.

Update (Actualizar): Permite la modificación de los registros existentes en la base de datos.

Delete (Eliminar): Permite la eliminación de registros existentes en la base de datos.

Miremos como conectarnos a sqlite3 con una base de datos en Python:

```python
# se recomienda leer https://docs.python.org/es/3/library/sqlite3.html
```

```python
import sqlite3

def crearbase():
    #de esta forma creo la base, si ya existe la lee
    con= sqlite3.connect('base_nombres_apellidos.db')
    #sqlite3.connect(ruta) devuelve un objeto de conexión, que a su vez es devuelto por crearbase().
    # Este objeto de conexión se puede utilizar para ejecutar consultas en una base de datos SQLite.

    return con

con=crearbase()
con
```

```python
def creartabla(con):
    #con cursor se puede agregar informaciÓn a la base de datos, es equivalente a un iterador pero dentro de una base de datos
    cursor= con.cursor()
    sql="CREATE TABLE IF NOT EXISTS tabla_identidad(Nombre text, Apellido text, DNI integer NOT NULL)"

    #sql="CREATE TABLE IF NOT EXISTS tabla_identidad(Nombre text, Apellido text, DNI integer PRIMARY KEY)"

    #sql="CREATE TABLE IF NOT EXISTS tabla_identidad(Nombre text, Apellido text, DNI integer, primary key (apellido, dni))"

    cursor=cursor.execute(sql)
    con.commit()

creartabla(con)

#DDL (Data Definition Language)
```

```python
def insertar_datos(nombre,apellido,dni):
    #con cursor se puede agregar informacion a la base de datos
    cursor= con.cursor()
    nombre=nombre.upper()
    apellido=apellido.upper()


    data=(nombre,apellido,dni)
    sql="INSERT INTO tabla_identidad(Nombre, Apellido, DNI) VALUES(?, ?, ?)"
    cursor=cursor.execute(sql, data)
    con.commit()

    # DML (Data Manipulation Language)
```

```python
insertar_datos("jorge", "ruis", 9880)
```

```python
insertar_datos("JUAN", "perez",9870 )
```

```python
def actualizar_apellido(apellido, apellido_modificado):
    #con cursor se puede modificar informacion a la base de datos
    cursor= con.cursor()

    apellido=str(apellido).upper()
    apellido_modificado=str(apellido_modificado).upper()
    # el orden dentro de la variable data debe tener el orden que primero aparece en la orden sql
    data=(apellido_modificado, apellido)
    sql="update  tabla_identidad set Apellido= ? where apellido= ?"
    cursor=cursor.execute(sql, data)
    con.commit()

    #DML (Data Manipulation Language)
```

```python
actualizar_apellido("RUIS", "RUIZ")
```

```python
def borrar_usuario(dni):


    cursor= con.cursor()


    data=(dni,)
    sql="DELETE from tabla_identidad where DNI= ?"
    cursor=cursor.execute(sql, data)

    con.commit()

# DML (Data Manipulation Language)
```

```python
borrar_usuario(9880)
```

```python
def execute_read_query(query):
    cursor= con.cursor()


    cursor.execute(query)
    result = cursor.fetchall()
    return result
```

```python
elegir_usuarios = "SELECT * from tabla_identidad"
users = execute_read_query(elegir_usuarios)

for user in users:
    print(user)


# DML (Data Manipulation Language)
```

Al ser el DNI un valor único, no debería ser posible la duplicación de información en ese campo. Podemos hacer lo siguiente:

```python
def creartabla(con):
    #Hacemos la nueva tabla con su respectivo PK pero primero borramos la anterior
    cursor= con.cursor()
    cursor.execute("DROP TABLE tabla_identidad")
    cursor.execute("CREATE TABLE tabla_identidad(Nombre text, Apellido text, DNI integer PRIMARY KEY)")
    con.commit()

creartabla(con)



#DDL (Data Definition Language)
```
